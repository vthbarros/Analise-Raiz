# Análise Gráfica da Raizen

## Importamos as libs necessárias

```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import mplfinance as mpf
import datetime as dt
```

## Definindo

```python
raizen = yf.download("RAIZ4.SA") # É feito o download dos preços da Raizen
raizen.info
trend_data = raizen['Close'].rolling(window=20).mean() # Criamos uma linha de tendência em 20 períodos para acompanhar no gráfico
print(raizen)
```

```python
mpf.plot(raizen,
         type="candle",
         addplot=mpf.make_addplot(trend_data, color='r'), # Coloca a linha de tendência definida antes
         mav=(3), #define a média móvel
         style="yahoo", 
         volume=True,
         title="Gráfico da RAIZ4.SA em Candlestick usando Python",
         ylabel="Preços",
         ylabel_lower="Volume",
         xlabel="Data",
         figsize=(20,5)) #define o tamanho do gráfico gerado, sendo (Largura, Altura)
```
